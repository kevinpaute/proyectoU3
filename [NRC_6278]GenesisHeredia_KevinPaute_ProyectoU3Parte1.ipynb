# Cargar dataset

```python
#Se cargan las imagenes del dataset colocando la ruta de la carpeta general
ruta_imagenes = "dataset_imagenes"
#Función que permite listar las carpetas que contiene 
clases = os.listdir(ruta_imagenes)
#Función que permite clasificar los nombres de cada carpeta
clases.sort()
#Imprimir los nombres de las carpetas
print(clases)
```

# Preprocesamiento de imágenes

```python
'''
    Ciclo for para copiar las imágenes y realizar data aumented
'''
for n_clase in clases:
    #Selecciona la ruta de las imagenes y las une con el nombre de la calse
    cur_path = ruta_imagenes + "/" + n_clase
    elements = os.listdir(cur_path)
    
    for element in elements:
        #Muestra el bucle de las imagenes que está aplicandose data augmented
        print("Copiando imagen - {} en dataset_imagenes_augmented/{}".format(element, n_clase))
        # lee el origen de la ruta
        src = os.path.join(ruta_imagenes + "/" + n_clase, element) 
        # Guarda las imagenes en la carpeta especificada
        dst = os.path.join("dataset_imagenes_augmented/" + n_clase, element) 
        shutil.copy(src, dst)
```

## Data augmentation

```python
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.4,
        horizontal_flip=True,
        fill_mode='nearest')
```

## Filtro NLM para eliminación de ruido de imágenes

```python
#Se cargan las imagenes del dataset colocando la ruta de la carpeta general
os.chdir(r"C:\Users\usuar\Documents\PROYECTOIAU3\proyectoU3")
ruta_imagenes = "dataset_imagenes_augmented"
#Función que permite listar las carpetas que contiene 
nombre_clases = os.listdir(ruta_imagenes)
#Función que permite clasificar los nombres de cada carpeta
nombre_clases.sort()
#Imprimir los nombres de las carpetas
print(nombre_clases)
```

```python
#Imprime el numero de imágenes por carpeta
for nombre_clase in nombre_clases:
    print("La Clase "+nombre_clase+ " tiene "+str(len(os.listdir(ruta_imagenes + "/" + nombre_clase)))+" imágenes")

#Numero total de imagenes de todas las carpetas
print("\nEl dataset tiene un total de "+ str(sum(len(os.listdir(os.path.join(ruta_imagenes, nombre_clases[i]))) 
                                                 for i in range(len(nombre_clases)))) +" imágenes")
```

```python
for n_clase in nombre_clases:
    cur_path = ruta_imagenes + "/" + n_clase
    cur_label = n_clase
    i = 1

    for archivo in glob.glob(cur_path + "/*"):
        print("Procesando Imagen - {} en {}".format(i, cur_label))
        # Lee la ruta de las imagenes
        image = cv2.imread(archivo)
        # Cambia el ruido de las imagenes
        dst = cv2.fastNlMeansDenoisingColored(image,None,10,10,7,21)
        b,g,r = cv2.split(dst)
        rgb_dst = cv2.merge([r,g,b])
        
        save_path = 'dataset_imagenes_denoised/'+ cur_label
        cv2.imwrite(os.path.join(save_path , cur_label + '_' + str(i) + '.jpg'),
                    cv2.cvtColor(rgb_dst,cv2.COLOR_RGB2BGR))
        

        # Muestra la actualización de bucle
        i += 1
```

## Background Removal

```python
#Se cargan las imagenes del dataset colocando la ruta de la carpeta general
os.chdir(r"C:\Users\usuar\Documents\PROYECTOIAU3\proyectoU3")
ruta_imagenes = "dataset_imagenes_denoised"
#Función que permite listar las carpetas que contiene 
nombre_clases = os.listdir(ruta_imagenes)
#Función que permite clasificar los nombres de cada carpeta
nombre_clases.sort()
#Imprimir los nombres de las carpetas
print(nombre_clases)
```

```python
for n_clase in nombre_clases:
    cur_path = ruta_imagenes + "/" + n_clase
    cur_label = n_clase
    i = 1
    
    for archivo in glob.glob(cur_path + "/*"):
        print("Procesando Imagen - {} en {}".format(i, cur_label))
        # Leer la ruta de imagenes
        image = cv2.imread(archivo)
        # Eliminar el fondo de la imagen
        output = remove(image)
        # Guarda las imagenes en la ruta removedBG
        save_path = 'dataset_imagenes_removedBG/'+ cur_label
        cv2.imwrite(os.path.join(save_path , cur_label + '_' + str(i) + '.png'), output)
        

        # Muestra la actualización de bucle
        i += 1
```
